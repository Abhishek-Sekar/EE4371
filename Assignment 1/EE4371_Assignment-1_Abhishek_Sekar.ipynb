### (a)Write a Python function that takes a positive integer n, and returns the sum of the squares of all the positive integers smaller than n.  

```python
import time

def square_sum(n): #function definition: input, a positive integer n
    if(n <= 0):
        return -1 #error if you input a negative number
    elif(type(n) != int):
        return -1 #error if you don't input an integer
    else:
        s = 0          # variable which will store the answer
        for i in range(1,n): # for loop from 1 to n, n not included 
            s += i**2         #keeps adding the i^th square to s all the way upto the integer right below n
        return s           # returns the sum of squares of all the positive integers lesser than n
    
def alter_square_sum(n):  #alternate way to compute the sum of squares without wasting iterations
    if(n <= 0):
        return -1 #error if you input a negative number
    elif(type(n) != int):
        return -1 #error if you don't input an integer
    else:
        return int(n*(n-1)*(2*n - 1)/6) # returns the answer using a known result for sequences
    

print("Please input a positive integer.")
n = int(input())
start = time.time()
print('The sum of squares of all the positive integers below n is',square_sum(n)) #output through square_sum
end = time.time()
print(end-start)
start_1 = time.time()
print('The sum of squares of all the positive integers below n is',alter_square_sum(n)) #output through alter_square_sum
end_1 = time.time()
print(round(end_1 - start_1,8))
```

### (b)	Write a Python function that takes a positive integer n, and returns the sum of the squares of all the odd positive integers smaller than n.  

```python
def odd_square_sum(n):  #function definition: input, a positive integer n
    if(n <= 0):         #error if a negative number is being input 
        return -1
    elif(type(n) != int):
        return -1       #error if you don't input an integer
    else:
        s = 0          #sum variable is initialised as in the previous case
        for i in range(1,n,2):   #increment in steps of 2 this time, as we only need odd numbers(2k+1)<n for appropriate k(k < (n-1)/2)
            s += i**2      #update sum variable for each odd number
        return s           #returns the final sum

def odd_alter_square_sum(n):  #alternate way to compute the sum of squares without wasting iterations
    if(n <= 0):
        return -1 #error if you input a negative number
    elif(type(n) != int):
        return -1 #error if you don't input an integer
    else:
        k = n//2 #total of n//2 odd integers below n ,n//2 is integer division
        return int(k*(2*k+1)*(2*k - 1)/3) # returns the answer using a known result for sequences
    
    
print("Please input a positive integer.")
n = int(input())

print('The sum of squares of all the odd positive integers below n is',odd_square_sum(n))
print('The sum of squares of all the odd positive integers below n is',odd_alter_square_sum(n))
    
```

### Write a Python function that takes a sequence of integer values and determines if there is a distinct pair of numbers in the sequence whose product is odd. 

```python
def prodd(n): #function definition: inputs a list which holds a sequence of integer values, returns a string 'Yes' or "No"
    count = 0
    for i in range(len(n)):
        count = 0       #count initialized to 0
        if(n[i] % 2 == 1):
            count += 1    #update count if odd integer
            for j in range(i+1,len(n)): #iterate through new index
                if(n[j] % 2 == 1):    # check if new integer is odd
                    if(n[j] != n[i]): #check if the new odd integer is unique
                        count += 1    #update count
                        break
        if(count == 2):
            return "Yes"
    return 'No'

print(prodd([-3,2,3,4,3,3,4]))
print(prodd([1,2,3,4,5]))
print(prodd([2,4,6,1]))
print(prodd([2,4,6,8]))
```

## Q4
### Write a Python function that counts the number of vowels in a given character string. 

```python
def vowel_count(s):
    s = s.lower() #converting all characters in the string to lowercase
    count = 0
    for i in range(len(s)):
        if(s[i] == 'a'):
            count += 1
        elif(s[i] == 'e'):
            count += 1
        elif(s[i] == 'i'):
            count += 1
        elif(s[i] == 'o'):
            count += 1
        elif(s[i] == 'u'):
            count += 1
    return count

print(vowel_count('Abhishek'))
print(vowel_count('Algorithms'))
print(vowel_count('Aaauei'))
print(vowel_count('Myth'))
```

### Write a Python program that outputs all possible strings formed by using the characters ‘_c_’, ‘_a_’, ‘_t_’, ‘_d_’, ‘_o_’, and ‘_g_’ exactly once.

```python
 def all_possible_strings(input_str, start_ind = 0):
    if (start_ind == len(input_str)-1):
        # we've gotten to the end, print the permutation
        print (" ".join(input_str)) #joining the characters of the list in a space separated manner
    for i in range(start_ind, len(input_str)):
        # copy the input string into a list str_copy
        str_copy = [c for c in input_str] # copy of input_str
        # swap the current index of str_copy with the start_ind given
        str_copy[start_ind], str_copy[i] =str_copy[i], str_copy[start_ind]
        # recurse on the portion of the stringthat has not been swapped yet
        all_possible_strings(str_copy,start_ind + 1)
        
all_possible_strings('catdog')
```

### Write a Python program that can take a positive integer greater than 2 as input and write out the number of times one must repeatedly divide this number by 2 before getting a value less than 2.   

```python
n = int(input())
start = time.time()
i = 0
while(n >= 2):
    n = n/2
    i += 1
    
print("Using iterative approach",i)

end = time.time()
print(end - start)
    
```
